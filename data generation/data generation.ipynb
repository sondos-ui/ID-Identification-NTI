```python
import pandas as pd
import convert_numbers
import random
import os
```

```python
df = pd.read_csv("Arabic_names.csv")
```

```python
df.head()
```

```python
women_names = list(df[df['Gender']=='F']['Name'])
men_names = list(df[df['Gender']=='M']['Name'])
```

```python
all_ids=[]
```

```python
def generate_arabic_ids(num): #number of IDs we want to generate
    ids=[]
    spaces=[18,15,12,9,6,2]
    while (len(ids) != num):
        listOfNumbers=[]
        
        for x in range (0, 14):
            listOfNumbers.append(str(random.randint(0, 9)))
            
        for i in spaces:
            listOfNumbers.insert(i,' ')
            
        
        temp_id ="".join(n for n in listOfNumbers)
        temp_id = convert_numbers.english_to_hindi(temp_id)
        
        if temp_id not in all_ids:
            all_ids.append(temp_id)
            temp_id=list(temp_id)
            final_ids=[0 for x in range(14+6)]
            j=0
            for i in range(20):
                if i in spaces:
                    final_ids[i]=" "
                elif j<=len(temp_id): 
                    final_ids[i]=temp_id[j]
                    j+=1
                
            ids.append(" ".join(final_ids))
    
    return ids
```

```python
def generate_last_names(num, names):
    last_names = []
    
    for i in range(num):
        last_name = random.choice(names)+" "+random.choice(names)+" "+random.choice(names)
        last_names.append(last_name)
    
    return last_names    
```

```python
def generate_first_names(num, names): # num: Number needed to be generated. names: names we'll choose from
    first_names = []
    
    for i in range(num):
        first_name = random.choice(names)
        first_names.append(first_name)
    
    return first_names
```

```python
def generate_data(img_path,num,df,gender=''):
    
    l_names = generate_last_names(num, men_names)
    ar_id = generate_arabic_ids(num)
    faces=os.listdir(img_path)
    faces=[img_path+'/'+str(x) for x in faces]
   
    df['l_name']=l_names
    df['ssn']=ar_id
    df['face']=faces
    
    if gender=='M':
        df['gender'][0:num]='M'
        f_names = generate_first_names(num, men_names)
        
    else: 
        df['gender'][0:num]='F'
        f_names = generate_first_names(num, women_names)
        
    df['f_name']=f_names
        
    
    return df
    
```

```python
female_data=pd.DataFrame(columns=['f_name','l_name','gender','face','ssn'])
female_data=generate_data('women',530,female_data,'F')

female_data.head()
```

```python
male_data=pd.DataFrame(columns=['f_name','l_name','gender','face','ssn'])
male_data=generate_data('men',530,male_data,gender='M')

male_data.head()
```

```python
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
```

```python
def generate_card(data):
    
    font_txt = ImageFont.truetype("majallab.ttf", size=55)
    font_num = ImageFont.truetype("majallab.ttf", size=57)
    template = Image.open("template.jpg")
    pic = Image.open(data['face']).resize((250, 330))
    pic=pic.convert("L")

    template.paste(pic, (25, 42, 275, 372))
    draw = ImageDraw.Draw(template)
    
    direction = 'rtl' #Right to Left
    Alignment= 'right'
    
    text= data['f_name']
    f_name_wrapped = textwrap.fill(text, 100)
    
    text= data['l_name']
    l_name_wrapped = textwrap.fill(text, 100)

    # get text box size
    f_name_w, _ = draw.textsize(f_name_wrapped, font=font_txt, direction="rtl")
    l_name_w, _ = draw.textsize(l_name_wrapped, font=font_txt, direction="rtl")

    
    
    draw.text(((970- (f_name_w)), 160), f_name_wrapped, stroke_width=0,font=font_txt, fill='black',align=Alignment, direction=direction)
    draw.text(((970- (l_name_w)), 226), l_name_wrapped,stroke_width=0, font=font_txt, fill='black',align=Alignment, direction=direction)



    
    
    draw.text((454, 495), str(data['ssn']), stroke_width=1,font=font_num, fill='black',align=Alignment, direction=direction)
   
    return template
```
